```python
import torch
import torch.onnx
from torchvision import models

# Load the ResNet18 model without pre-trained weights
model = models.resnet18(pretrained=False)  # Load the architecture

# Change the final fully connected layer to match the number of classes in your model
num_features = model.fc.in_features
model.fc = torch.nn.Linear(num_features, 47)  # Set the number of output classes to 47

# Load the trained weights
model.load_state_dict(torch.load("plant_disease_model.pth"))
model.eval()

# Dummy input for tracing the model
dummy_input = torch.randn(1, 3, 224, 224)  # 1 image, 3 channels (RGB), 224x224 size

# Export the model to ONNX
torch.onnx.export(model, dummy_input, "plant_disease_model.onnx", verbose=True)
```
